```python
#Prompt template text summarization
#when we combine prompt template witl LLM we refer as LLMChain
```

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate

generic="""
write a summary of the following speech:
speech:{speech}
Translate the summary to {language}
"""
prompt=PromptTemplate(
 input_variables=['speech','language'],
 template=generic 
    
)
```

```python
prompt
```

```python
complete_prompt=prompt.format(speech=speech,language="hindi")
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain=LLMChain(llm=llm,prompt=prompt)
```

```python
summary=llm_chain.run({'speech':speech,'language':'hindi'})
```

```python
#stuff doc chain 
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("apjspeech.pdf")
docs=loader.load_and_split()
```

```python
#pages
```

```python
template=""" 
write a short and concise summary of the speech
speech:{text}

"""
prompt=PromptTemplate(input_variables=['text'],template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain 
#load summarize chain combines everything wrt text summarization
chain=load_summarize_chain(llm, chain_type='stuff',prompt=prompt, verbose=True)
```

```python
output_summary=chain.run(docs)
```

```python
output_summary
```

```python
#using map reduce method 
from langchain.text_splitter import RecursiveCharacterTextSplitter
final_docs=RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=50).split_documents(docs)
```

```python
#final_docs
```

```python
len(final_docs)
```

```python
prompt
```

```python
chunks_prompt="""
Please summarize the below speech:
Speech :{text}
Summary:
"""
map_prompt_template=PromptTemplate(input_variables=['text'],
                                   template=chunks_prompt)
```

```python
summary_chain=load_summarize_chain(llm=llm,chain_type="map_reduce",
                                   map_prompt=map_prompt_template)
```

```python
final_concat_prompt="""

Provide the final summary of the entire speech
with these important points.

Add a motivational title,
Start the precise summary with an introduction and provide the summary in 
number points.
Speech:{text}
"""
final_prompt_concat_template=PromptTemplate(input_variables=['text'], template=final_concat_prompt)
```

```python
summary_chain_final_concat=load_summarize_chain(llm=llm,
                                    chain_type="map_reduce",
                                   map_prompt=map_prompt_template,
                                    combine_prompt=final_prompt_concat_template,
                                               verbose=True)
```

```python
output=summary_chain_final_concat.run(final_docs)
```
